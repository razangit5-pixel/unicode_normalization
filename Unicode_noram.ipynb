# What is Encoding?
    Encoding is the process of converting text (human-readable characters) into bytes (machine-readable binary data).
## Why does it matter?
    Encoding is needed when:

    1. Saving text to a file
    2. Sending data over a network
    3. Working with APIs or databases
    4. Handling emojis or special characters (like Arabic! 'مرحبا'.encode('utf-8'))
    

```python
text = "I love pizza 😍 ! Shall we book a  to gizza? 🎍"
Text = text.encode("utf-8") #converts it into a **bytes object** using the UTF-8 encoding scheme
print(Text)
```
