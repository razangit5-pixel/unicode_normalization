## Correct Steps

### Import Modules

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
```

### request to access the html page 


```python
myurl = "https://stackoverflow.com/questions/37890898/how-to-set-env-variable-in-jupyter-notebook"

headers = {
    "User-Agent": "Mozilla/5.0"
}
req = Request(myurl, headers=headers)
```

### Retrieve the html content

```python
html = urlopen(req).read()
```

###  parse the html to a soup object which view it as a tree

without it, python will think the html is just a string



```python
soupified = BeautifulSoup(html, "html.parser")

```

### retrieve the title


```python
question = soupified.find("div", class_="question")
questiontext = question.find("div", class_="s-prose js-post-body")
print("\nQuestion:\n", questiontext.get_text(strip=True))
```

### retrieve the question


```python
question = soupified.find("div", class_="question")
questiontext = question.find("div", class_="s-prose js-post-body")
print("\nQuestion:\n", questiontext.get_text(strip=True))
```

### retrieve the best answer


```python
accepted = soupified.find("div", class_="accepted-answer")

if accepted:
    answertext = accepted.find("div", class_="s-prose js-post-body")
    print("\nBest Answer:\n", answertext.get_text(strip=True))
else:
    print("\nNo accepted answer found. Showing top voted answer instead.\n")
    
    first_answer = soupified.find("div", class_="answer")
    if first_answer:
        body = first_answer.find("div", class_="s-prose js-post-body")
        print(body.get_text(strip=True))
    else:
        print("No answers found.")


```
